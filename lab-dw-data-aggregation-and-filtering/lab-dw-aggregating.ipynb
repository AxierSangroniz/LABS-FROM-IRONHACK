```python
ordencolumnas2 = ['Unnamed: 0', 'Customer', 'State', 'Customer Lifetime Value',
       'Response', 'Coverage', 'Education', 'Effective To Date',
       'EmploymentStatus', 'Gender', 'Income', 'Location Code',
       'Marital Status', 'Monthly Premium Auto', 'Months Since Last Claim',
       'Months Since Policy Inception', 'Number of Open Complaints',
       'Number of Policies', 'Policy Type', 'Policy', 'Renew Offer Type',
       'Sales Channel', 'Total Claim Amount', 'Vehicle Class', 'Vehicle Size',
       'Vehicle Type']
for columna in ordencolumnas2:
    print(dfcopy[columna].value_counts())
```

```python
dfcopy = dfcopy.drop_duplicates(keep="first")
```

```python
dfcopy.columns = dfcopy.columns.str.lower().str.replace(" ", "_")
```

```python
dfcopy.drop("unnamed:_0", axis=1, inplace=True)
dfcopy.drop("vehicle_type", axis=1, inplace=True)
```

```python
dfcopy = dfcopy.dropna(how="all")
```

```python
dfcopy["state"].fillna("no_data", inplace=True)
dfcopy["response"].fillna("no_data", inplace=True)
dfcopy["months_since_last_claim"].fillna(dfcopy["months_since_last_claim"].mean(), inplace=True)
dfcopy["number_of_open_complaints"].fillna(dfcopy["number_of_open_complaints"].mean(), inplace=True)
dfcopy["vehicle_class"].fillna("no_data", inplace=True)
dfcopy["vehicle_size"].fillna(0, inplace=True)
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

Utilizando el DataFrame original, analiza el promedio del total_claim_amount para cada tipo de póliza y género de los clientes que han respondido "Sí" a la última campaña de marketing. Escribe tus conclusiones.

```python
dfresponse = dfcopy[dfcopy["response"] == "Yes"]
multiple = dfresponse.groupby(["policy_type", "gender"])
```

```python
df5 = multiple["total_claim_amount"].mean()
df5 = df5.reset_index()
df5
```

```python
df_pivot = pd.pivot_table(df5, values='total_claim_amount', index=['policy_type', "gender"], aggfunc="mean")
df_pivot
```
